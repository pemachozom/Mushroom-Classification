# Use of Pipeline

```python
x=data.drop('class', axis=1)
y=data['class']
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2, random_state=1, shuffle=True, stratify=y)
```

```python
print(X_test.shape)
print(X_train.shape)
```

Updating x_train and x_test with feature selected

```python
X_train = X_train[['cap-surface', 'bruises', 'gill-spacing', 'gill-size',
       'gill-color', 'stalk-root', 'ring-type', 'spore-print-color',
       'population', 'habitat']]
X_test = X_test[['cap-surface', 'bruises', 'gill-spacing', 'gill-size',
       'gill-color', 'stalk-root', 'ring-type', 'spore-print-color',
       'population', 'habitat']]
```

```python
cols = X_train.select_dtypes(include=['category']).columns
cols
```

```python
#  Building the column transformation pipeline
col_transformer = Pipeline(steps = 
                          [
                              ('impute', SimpleImputer(strategy='most_frequent')),
                              ('encode', OneHotEncoder(handle_unknown="ignore"))
                          ])
```

```python
# preprocessing = ColumnTransformer(transformers=
#                                  [
#                                      ('features', col_transformer, cols)
#                                  ])
```

Creating Pipeline Model

KNN Classification

```python
KNNestimator=KNeighborsClassifier(n_neighbors=3)
```

```python
KNNpipelinemodel = Pipeline(steps=[
    ('pre', col_transformer),
    ('est', KNNestimator)
])
```

```python
from sklearn import set_config
set_config(display='diagram')
KNNpipelinemodel.fit(X_train, Y_train)
```

```python
ypred = KNNpipelinemodel.predict(X_test)
```

```python
print("KNN Classifier report: \n\n", classification_report(Y_test, ypred))
```
